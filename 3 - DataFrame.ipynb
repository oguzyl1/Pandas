```python
import pandas as pd 
```

```python
#elimizde veri adında bir sözlük var 
veri = {"isim":["Ali","Can","Efe","Ayşe","Buse","Alp","Nur"],
        "puan":[90,80,85,75,95,60,65],
        "spor":["Güreş","Futbol","Kayak","Yüzme","Tenis","Karate","Sörf"],
        "cinsiyet":["E","E","E","K","K","E","K"]}

#bu sözlüğü bu şekilde bir dataFrame e dönüştürebiliriz
df = pd.DataFrame(veri)
df
```

```python
#ilk 5 veriyi görmek için
df.head()
#son 5 veriyi görmek için
df.tail()
#bunlar default olarak 5 er tane veriyor istediğimiz kadarını görmek istediğimiz zaman head(2) tail(20) şeklinde kullanarak görebiliriz
```

```python
#sütunların sırasını istediğimiz şekilde şekillendirebiliriz
df = pd.DataFrame(veri,columns=["isim","spor","puan","cinsiyet"])
df
```

```python
#dataFrame de olmayan bir kolon eklersek bunu ekler ve değerlerini NaN (eksik veri) olarak girer
df = pd.DataFrame(veri,columns=["isim","spor","puan","cinsiyet","yaş"])
df
```

```python
#istersek indexlere etiket atayabiliriz, index de 1 2 3 4 ... yerine bizim girdiğimiz değerler görünür 
df = pd.DataFrame(veri,columns=["isim","spor","puan","cinsiyet","yaş"]
                  ,index = ["bir","iki","uc","dort","bes","alti","yedi"])
df
```

```python
#dataFrame içindeki belirli bir sütunu alabiliriz 
df["spor"]
```

```python
#birden fazla sütunu almak istersek bir değişkene atayıp onun üzerinden çağırabiliriz
sutunlarim = ["isim","spor"]
df[sutunlarim]
```

```python
#istediğimiz sütunu aşağıdaki gibi de çağırabiliriz ama sütun isminde boşluk varsa hata alırız yukarıdaki gibi kullanmak zorunda kalırız
df.spor
# örn: sütunun adı Spor Bilimleri olsaydı df.Spor Bilimleri yazamazdık
```

```python
#belirli bir satırı seçmek istersek aşağıdaki işlemi yapabiliriz
df.loc[["bir"]]
```

```python
#birden fazla satırı almak istersek 
df.loc[["bir","yedi"]]
```

```python
#yaş sütununu eklemiştik ve o sütun boştu onu doldurmak için aşağıdaki işlemi yapabiliriz
deger = [17,18,19,17,19,20,18]
df["yaş"] = deger
df
```

```python
#sütun oluşturuken  mantıksal işlemlere göre sütunlar oluşturabiliriz
df["gecti"] = df.puan > 70
df
```

```python
#sütun silmek için
del df["gecti"]
df
```
