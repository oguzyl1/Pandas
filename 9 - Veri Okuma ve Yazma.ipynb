```python
import pandas as pd
```

dışarıdan aldığımız verileri programa aktarırken dikkat etmemiz gereken adımlar var. Örneği bize gelen dosyada veriler arasında bir tab boşluk varsa bu dosyayı read_table() fonksiyonu ile alırız virgüller varsa read_cvs() fonksiyonu ile alırız

```python
#örneğin
df = pd.read_table("DataSets\\data2.txt")
df
```

```python
#bu dosyada veriler , ile ayrılmış biz bunu table ile çağırdığımız için tek bir veri olarak göründü cvs olarak çağırsaydık 
df2 = pd.read_csv("DataSets\\data2.txt")
df2
```

```python
#bu dosyayı table olarak çağırmak istediğimizi varsayalım bu durumda aşağıdaki parametreyi kullanmamız lazım 
df = pd.read_table("DataSets\\data2.txt",sep=",")
df
```

```python
#bize gelen her dosyada sütun ismi olmayabilir bu dosyada olduğu gibi bizim bunu belirtmemiz lazım
df = pd.read_table("DataSets\\data2.txt",sep=",",header=None)
df 
```

```python
#istersek kendimiz sütun isimleri verebiliriz
df = pd.read_table("DataSets\\data2.txt",sep=",",header=None,names=["isim","puan","cinsiyet"])
df
```

```python
#farzedelim ki isim sütununu index olarak kullanmak istiyoruz 
df = pd.read_table("DataSets\\data2.txt",sep=",",header=None,names=["isim","puan","cinsiyet"],
                   index_col="isim")
df
```

```python
#veri seti çoklu sütünlardan oluşuyorsa ve bizde bu veri setini aşamalı bir biçime getirmek istersek aşağıdaki işlemleri yapmalıyız
df3 = pd.read_table("DataSets\\data3.txt",sep=",")
df3
```

```python
df3 = pd.read_table("DataSets\\data3.txt",sep=",",
                    index_col=["lesson","name"])
df3
```

```python
#aşağıdaki veri setinde yorum satırları bulunmakta ve bu satırların okunmasını istemiyoruz
df4 = pd.read_table("DataSets\\data4.txt",sep=",")
df4
```

```python
#okutmak istemediğimiz satırların indexlerini aşağıdaki fonksiyona girerek bu satırları atlayabiliriz
df4 = pd.read_table("DataSets\\data4.txt",sep=",",
                    skiprows=[0,2])
df4
```

```python
#veri setindeki belli kolanları okutmak istiyorsak
df4 = pd.read_table("DataSets\\data4.txt",sep=",",
                    skiprows=[0,2],
                    usecols=[0,1])
df4
```

```python
#elimizdeki veri setinde eksik veri olup olmadığını öğrenmek için aşağıdaki tabloyu ekleyelim
df5 = pd.read_table("DataSets\\data5.txt",sep=",")
df5
```

```python
#tabloda eksik veri varsa true boş değilse false dönecek 
pd.isnull(df5)
```

```python
#elimizdeki tablodaki verilerde cinsiyette Tom ve score da -5 var bunlar yanlış veriler biz bunu eksik veri olarak değiştirmek istersek
eksik_veri = {"score":[-5],"sex":["Tom","n.a"]}
#eksik veri olarak işaretlemek istediğimiz kolonları ve verileri sözlük veri yapısı ile belirttik
#aslında direkt na_values = ["Tom"] da diyebilirdik ama bu durumda name sütunundaki Tom da NaN olacaktı bunun için bu şekilde belirttik
df5 = pd.read_table("DataSets\\data5.txt",sep=",",na_values=eksik_veri)
df5
```

```python
#verideki değerleri ayırmak için virgül yerine farklı bir sembol kullanmak istersek 
import sys
df6 = pd.read_table("DataSets\\data.txt")
df6
```

```python
df6.to_csv(sys.stdout,sep="|")
```

```python
#elimizdeki eksik verilerin NaN yerine kendi istediğimiz şekilde görünmesinin istiyorsak 
df6 = pd.read_table("DataSets\\data.txt")
df6.to_csv(sys.stdout,sep="|",
           na_rep="NULL")
```
