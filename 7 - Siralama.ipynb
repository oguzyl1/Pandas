```python
import pandas as pd ; import numpy as np
```

```python
#ilk olarak series veri yapısında sıralama işlemine bakacağız
s = pd.Series(range(5),index=["e","d","a","b","c"])
s
```

```python
#indexe göre sıralama , a-z veya 1-...
s.sort_index()
```

```python
#dataFrame veri yapısındaki değişkenin sıralaması 
df = pd.DataFrame(np.arange(12).reshape(3,4),index=["iki","bir","uc"],columns=["d","a","b","c"])
df
```

```python
df.sort_index()
```

```python
#eğer sütunlara göre bir sıralama yapmak istersek axis = 1 ifadesi kullanılmalı 
df.sort_index(axis=1)
```

```python
#sıralamayı azalan sıralamada (z - a , .... - 1 ) yapmak istersek "ascending=False" parametresi kullanılmalıdır  
df.sort_index(axis=1,ascending=False)
```

```python
#indexlere göre değil de tablodaki değerlere göre işlem yapmak istersek sort_values fonksiyonu kullanılır
s2 = pd.Series([5,np.nan,3,-1,9])
s2
```

```python
#eğer eksik veri var ise bu sona atılır 
s2.sort_values()
```

```python
#dataFrame için değerlere göre sıralama işlemi 
df2 = pd.DataFrame({"a":[5,3,-1,9],"b":[1,-2,0,5]})
df2
```

```python
#b sütununa göre değerlerin sıralaması 
df2.sort_values(by="b")
```

```python
#birden fazla sütunun aynı anda sıralanması için
df2.sort_values(by=["b","a"])
```
